```python
import torch
import os
import glob
import matplotlib.pyplot as plt

# Folder where checkpoints are stored
folder = 'TranAD/checkpoints/TranAD_NAB/'  # Replace with your path

# Find all checkpoint files for different learning rates
ckpt_files = glob.glob(os.path.join(folder, 'model_lr_*.ckpt'))

# Dictionary to hold results
lr_results = {}

# Load each checkpoint
for ckpt_file in ckpt_files:
    # Extract learning rate from filename
    base = os.path.basename(ckpt_file)
    lr_str = base.split('_lr_')[-1].replace('.ckpt', '')
    lr = float(lr_str)

    # Load checkpoint
    checkpoint = torch.load(ckpt_file, map_location='cpu')
    accuracy_list = checkpoint['accuracy_list']  # List of (loss, accuracy) tuples

    # Store accuracies only
    losses = [loss for loss, _ in accuracy_list]
    lr_results[lr] = losses

# Plotting
plt.figure(figsize=(10, 6))
for lr, acc_list in sorted(lr_results.items()):
    plt.plot(range(1, len(acc_list) + 1), acc_list, label=f'lr={lr}')

plt.xlabel('Epoch')
plt.ylabel('Average loss')
plt.title('Learning Rate Comparison NAB')
plt.legend(loc="upper right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
